```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
import scipy
```

```python
data = pd.read_csv('dataset/epa-sea-level.csv')
```

```python
data.isnull().sum()
```

```python
data
```

```python
fig1 , ax1 = plt.subplots(figsize= (10,6))

ax1.scatter(x=data['Year'], y=data['CSIRO Adjusted Sea Level'])


fig1.show()
```

```python
res = scipy.stats.linregress(x=data['Year'], y=data['CSIRO Adjusted Sea Level'])
```

```python
future = np.arange(2000,2050)

fig2, ax2 = plt.subplots(figsize=(10,6))

ax2.scatter(x=data['Year'], y=data['CSIRO Adjusted Sea Level'])
ax2.plot(data['Year'], res.intercept + res.slope*data['Year'], 'r')
ax2.plot(future, res.intercept + res.slope*future, 'g')

ax2.set_title('Sea Level')
ax2.set_xlabel('Year')
ax2.set_ylabel('Sea Level (inches)')
ax2.legend()

fig2.show()
```

```python
new_data = data[data['Year'] >= 2000]
```

```python
new_res = scipy.stats.linregress(x=new_data['Year'], y=new_data['CSIRO Adjusted Sea Level'])
```

```python
fig3, ax3 = plt.subplots(figsize=(10,6))

ax3.scatter(new_data['Year'], new_data['CSIRO Adjusted Sea Level'])
ax3.plot(future, new_res.intercept + new_res.slope*future, 'g')

ax3.set_title('Sea Level')
ax3.set_xlabel('Year')
ax3.set_ylabel('Sea Level (inches)')
ax3.legend()

fig3.show()
```
